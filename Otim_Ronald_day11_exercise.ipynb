```python
import pandas as pd

# Load your dataset into a pandas DataFrame
df = pd.read_csv('heart.csv.xls')
```

```python
# Display the first few rows of the dataframe
print(df.head())
```

```python
# Display the summary statistics of the dataframe
print(df.describe())
```

```python
# Display the information about the dataframe
print(df.info())
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print('Number of duplicate rows:', duplicates)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print('Missing values in each column:')
print(missing_values)
```

```python
# Display the columns of the dataframe
print(df.columns)
```

```python
# Remove duplicates
original_count = len(df)
df = df.drop_duplicates()
new_count = len(df)
```

```python
# Print the number of rows before and after removing duplicates
print('Original number of rows:', original_count)
print('Number of rows after removing duplicates:', new_count)
```

```python
# Calculate descriptive statistics
# Measures of central tendency
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0]
```

```python
# Measures of spread
range_values = df.max() - df.min()
variance_values = df.var()
std_deviation_values = df.std()
interquartile_range_values = df.quantile(0.75) - df.quantile(0.25)
```

```python
# Skewness and Kurtosis
skewness_values = df.skew()
kurtosis_values = df.kurtosis()
```

```python
# Print the calculated statistics
print('Mean values:\
', mean_values)
print('Median values:\
', median_values)
print('Mode values:\
', mode_values)
print('Range values:\
', range_values)
print('Variance values:\
', variance_values)
print('Standard Deviation values:\
', std_deviation_values)
print('Interquartile Range values:\
', interquartile_range_values)
print('Skewness values:\
', skewness_values)
print('Kurtosis values:\
', kurtosis_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plot histograms for each variable
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(15, 20))
fig.subplots_adjust(hspace=0.5, wspace=0.5)
axes = axes.ravel()

for idx, column in enumerate(df.columns):
    sns.histplot(df[column], ax=axes[idx], kde=True, color='blue')
    axes[idx].set_title(column + ' Distribution')
    axes[idx].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Plot box plots for each variable
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(15, 20))
fig.subplots_adjust(hspace=0.5, wspace=0.5)
axes = axes.ravel()

for idx, column in enumerate(df.columns):
    sns.boxplot(y=df[column], ax=axes[idx], color='green')
    axes[idx].set_title(column + ' Box Plot')
    axes[idx].set_ylabel('Values')

plt.tight_layout()
plt.show()
```

```python
# Scatter plots for age vs cholesterol and age vs max heart rate
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))

sns.scatterplot(x='age', y='chol', data=df, ax=axes[0], color='purple')
axes[0].set_title('Age vs Cholesterol')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Cholesterol')

sns.scatterplot(x='age', y='thalach', data=df, ax=axes[1], color='orange')
axes[1].set_title('Age vs Max Heart Rate')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Max Heart Rate')

plt.tight_layout()
plt.show()
```
